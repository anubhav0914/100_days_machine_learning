```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_trest ,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size =0.2)
```

```python
X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
#  here the categories decide the  precedence or the value or high value is given to the good then average and then poor and similarly for the education column
oe = OrdinalEncoder(categories = [['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
```

```python
X_test = oe.transform(X_trest)
```
