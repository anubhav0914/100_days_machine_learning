```python
import os
import subprocess
import tarfile
from typing import List

# Step 1: Clone Whisper repository and set up environment
if not os.path.exists("whisper"):
    os.system("git clone https://github.com/openai/whisper.git")

# Step 2: Download and extract Kathbath dataset
if not os.path.exists("Kathbath"):
    os.makedirs("Kathbath")
    os.system("wget https://github.com/AI4Bharat/vistaar/releases/download/v0.2/kathbath-hindi.tar.gz -P Kathbath/")
    with tarfile.open("Kathbath/kathbath-hindi.tar.gz", "r:gz") as tar:
        tar.extractall("Kathbath")

# Step 3: Download ASR evaluation tools
if not os.path.exists("asr-evaluation"):
    os.system("git clone https://github.com/belambert/asr-evaluation.git")

# Step 4: Download and extract test dataset
if not os.path.exists("test_dataset"):
    os.makedirs("test_dataset")
    os.system("wget https://asr.iitm.ac.in/Gramvaani/NEW/GV_Eval_3h.tar.gz -P test_dataset/")
    with tarfile.open("test_dataset/GV_Eval_3h.tar.gz", "r:gz") as tar:
        tar.extractall("test_dataset")

# Step 5: Run Whisper ASR model on test dataset
def run_whisper_asr(input_dir: str, output_dir: str):
    for audio_file in os.listdir(input_dir):
        if audio_file.endswith(".wav"):
            input_path = os.path.join(input_dir, audio_file)
            output_path = os.path.join(output_dir, os.path.splitext(audio_file)[0] + ".txt")
            subprocess.run(["python", "whisper/recognize.py", input_path, output_path])

# Step 6: Calculate WER using ASR evaluation tools
def calculate_wer(ground_truth: str, asr_output: str) -> float:
    result = subprocess.run(["python", "asr-evaluation/compute-wer.py", ground_truth, asr_output], capture_output=True, text=True)
    wer = float(result.stdout.strip().split()[-1])
    return wer

# Step 7: Run Whisper ASR on test dataset and calculate WER
input_dir = "test_dataset/GV_Eval_3h"
output_dir = "asr_output"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

run_whisper_asr(input_dir, output_dir)

wer_scores = []
for audio_file in os.listdir(input_dir):
    if audio_file.endswith(".txt"):
        asr_output = os.path.join(output_dir, audio_file)
        ground_truth = os.path.join(input_dir, os.path.splitext(audio_file)[0] + ".txt")
        wer = calculate_wer(ground_truth, asr_output)
        wer_scores.append(wer)

average_wer = sum(wer_scores) / len(wer_scores)
print("Average WER:", average_wer)

# Step 8: Report your findings
print("Implementation complete. Report your findings.")
```
