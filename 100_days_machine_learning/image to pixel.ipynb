```python
import cv2
import numpy as np
import pandas as pd

# Read the image
image = cv2.imread('new_face.jpeg')

# Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Get the dimensions of the image
height, width = gray_image.shape

# Define the patch size
patch_size = 20

# Initialize lists to store pixel data
pixels = []

# Iterate over each patch in the image
for y in range(0, height, patch_size):
    for x in range(0, width, patch_size):
        # Get the patch
        patch = gray_image[y:y+patch_size, x:x+patch_size]
        # Resize the patch to 20x20 pixels
        resized_patch = cv2.resize(patch, (20, 20))
        # Flatten the resized patch and append to the list
        pixels.append(resized_patch.flatten())

# Convert the pixel data to a DataFrame
df = pd.DataFrame(pixels)

# Save the pixel data to a CSV file
df.to_csv('image_patches.csv', index=False)

print("Patch data saved to image_patches.csv")
```
