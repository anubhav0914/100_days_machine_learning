### 3. Opening a csv file from an URL

```python


import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)

```

## 9.SkipRow /nrow Parameter 

```python
## to skip the row or column 
pd.read_csv('align_011.csv',skiprows=[1,2,3,4])

# will get the n rows 
pd.read_csv('align_011.csv',nrows=100)
```

## Encoding parameter 

```python
## if encoding utf != 8 that means the coding contains emojies or special type od language 
pd.read_csv('align_011.csv',encoding='latin-1')
```

## Skip bad lines 

```python
# if the data set have some lines like any row have extra column so we can use 

#depericated  
# pd.read_csv('align_011.csv',error_bad_lines=False)
pd.read_csv('align_011.csv',on_bad_lines="skip")
```

## dtypes parameter 

```python
pd.read_csv('align_011.csv',dtype={'5.7':float}).info()
```

## Hnadling dates 

```python
# intially the date we get from the dataset is in the form of object so we have to cahnge it into date format

pd.read_csv('align_011.csv',parse_dates=['5.7'])
```

## Converters 

```python
#when we need to covert the naem as there sort form so ....
def rename(name):
    if name == "required name":
       return 'fuck'
    else :
       return name 
        
pd.read_csv('align_011.csv',converters ={'5.7':rename})
```

## na_values (missing values) parameter 

```python
# when we want to consider a specific value as a  null value so ...
pd.read_csv('align_011.csv',na_values=['Thanyou'])
```
