```python
# MinMax Scalling 
# range will be 0-1
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('wine_data.csv/wine_data.csv',header = None ,usecols =[0,1,2])
df.columns=['class lable','alchol','malic acid']
```

```python
df= df.iloc[1:]
df = df.reset_index(drop =True)
```

```python
df
```

```python
sns.kdeplot(pd.to_numeric(df['malic acid']))
```

```python
sns.kdeplot(pd.to_numeric(df['alchol']))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the color palette dictionary with all unique values in 'class label' column as keys
color_dict = {1: 'red', 3: 'blue', 2: 'green'}

# Plot the scatterplot with hue
sns.scatterplot(data=df, x='malic acid', y='alchol', hue='class lable', palette=color_dict)

# Display the plot
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train ,y_test = train_test_split(df.drop('class lable',axis = 1 ), df['class lable'],test_size =0.3, random_state = 0)

X_train.shape ,X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)

# transform train and test sets 
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled =pd.DataFrame(X_train_scaled ,columns =X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled ,columns =X_test.columns)
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Plot scatterplot before scaling
sns.scatterplot(data=X_train, x='alchol', y='malic acid', hue=y_train, ax=ax1)
ax1.set_title('Before Scaling')

# Plot scatterplot after scaling
sns.scatterplot(data=X_train_scaled, x='alchol', y='malic acid', hue=y_train, ax=ax2)
ax2.set_title('After Scaling')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Convert 'alcohol' and 'malic acid' columns to numeric
X_train['alchol'] = pd.to_numeric(X_train['alchol'])
X_train['malic acid'] = pd.to_numeric(X_train['malic acid'])

# Create subplots
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Plot KDE plot before scaling
sns.kdeplot(data=X_train['alchol'], ax=ax1)
sns.kdeplot(data=X_train['malic acid'], ax=ax1)
ax1.set_title('Before Scaling')

# Plot KDE plot after scaling
sns.kdeplot(data=X_train_scaled['alchol'], ax=ax2)
sns.kdeplot(data=X_train_scaled['malic acid'], ax=ax2)
ax2.set_title('After Scaling')

# Display the plot
plt.show()
```

```python
fig ,(ax1,ax2) = plt.subplots(ncols =2 ,figsize =( 12,5))

#before scaling 
ax1.set_title('Alchol Distribution Before Scaling ')
sns.kdeplot(X_train['alchol'],ax =ax1)

# after scaling 
ax2.set_title('alchol daistribution after the standard scaling')
sns.kdeplot(X_train_scaled['alchol'] ,ax = ax2)
plt.show()
```
