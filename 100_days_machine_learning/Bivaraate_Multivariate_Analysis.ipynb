##  Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x='total_bill',y='tip',data=tips) # this is for 2 two colun  
```

```python
# let we need to konw that whixh coustomer is male and which one is female
sns.scatterplot(x='total_bill',y='tip',data=tips,hue='sex')
```

## Bar plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass',y='Fare',data=titanic)
```

```python
# IF WE NEEED TO KNOW ON THE BASIS OF THE GENDER
sns.barplot(x='Pclass',y='Age',data=titanic,hue='Sex')
```

## BOX PLOT (NUMERICAL -CATEGORICAL)

```python
sns.boxplot(x='Sex',y='Age',data=titanic,hue='Survived')
```

## Distplot(NUmerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False) # who not survived
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False) # who survived

```

## 5.HeatMap(Categorical =Categorical)

```python
# lets assume we have to find how many people die and alive in each pclass

sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# percentage of survival 
titanic.groupby('Pclass').mean()['Survived']*100
```

# clussterMap (Categorical - Categorical)


```python
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python

sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```
