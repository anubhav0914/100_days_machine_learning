```python
from bs4 import BeautifulSoup
import requests

# Loop through the pages
for i in range(0, 7):
    # Skip the page if i is 5
    if i != 5:
        # Request the webpage
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
        webpage = requests.get(
            f'https://www.atpeducation.com/Science-notes-for-class-9th-hindi-medium-chapter-8.+%E0%A4%97%E0%A4%A4%E0%A4%BF-{i}', headers=headers).text

        # Create BeautifulSoup object
        soup = BeautifulSoup(webpage, 'lxml')

        # Extract text from div with class "page-cover"
        text = ''
        for div in soup.find_all('div', class_="page-cover"):
            text += div.text + '\n'

        # Save the text to a text file
        with open(f'class-9th_gati.txt', 'a', encoding='utf-8') as f:
            f.write(text)

print("Text saved to class-9th_gati.txt")
```
