### Precision  : 
* Precision is one indicator of a machine learning model's performance – the quality of a positive prediction made by the model.
*  Precision refers to the number of true positives divided by the total number of positive predictions (i.e., the number of true positives plus the number of false positives).
*  formula : `True positive /(True positive + False positive)`

### Recall :

* Recall is a metric that measures how often a machine learning model correctly identifies positive instances (true positives) from all the actual positive samples in the dataset.
* You can calculate recall by dividing the number of true positives by the number of positive instances.
* formula :`True positive /(True positive + False Negative)`

### F1 score :
* The F1 score or F-measure is described as the harmonic mean of the precision and recall of a classification model.
* The two metrics contribute equally to the score, ensuring that the F1 metric correctly indicates the reliability of a model.
* FORMULA : `(2*Precision*Recall)/(Precision + Recall)`
  

```python
import pandas as pd 
```

```python
df = pd.read_csv('heart_disease_uci/heart.csv')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.iloc[:,0:-1],df.iloc[:,-1],test_size = 0.2 ,random_state= 2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
log = LogisticRegression()
dec = DecisionTreeClassifier()
```

```python
log.fit(X_train,y_train)
dec.fit(X_train,y_train)
```

```python
y_pred_log = log.predict(X_test)
y_pred_dec =dec.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score ,confusion_matrix
print(f'Accracy score logistic : {accuracy_score(y_test,y_pred_log):.2f}%' )
print(f'Accuracy score decession tree  :{accuracy_score(y_test,y_pred_dec):.2f}%')
```

```python
from sklearn.metrics import recall_score,precision_score,f1_score
```

```python
print('for logistic regression model ')
print('-'*50)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pred_log),columns= list(range(0,2)))
print(cdf)
print('-'*50)
print(f'precision : {precision_score(y_test,y_pred_log)}')
print(f'recall : {recall_score(y_test,y_pred_log)}')
print(f'F1 score {f1_score(y_test,y_pred_log)}')
```

```python
print('for desiccion tree model ')
print('-'*50)
cdf = pd.DataFrame(confusion_matrix(y_test,y_pred_dec),columns= list(range(0,2)))
print(cdf)
print('-'*50)
print(f'precision : {precision_score(y_test,y_pred_dec)}')
print(f'recall : {recall_score(y_test,y_pred_dec)}')
print(f'F1 score {f1_score(y_test,y_pred_dec)}')
```
