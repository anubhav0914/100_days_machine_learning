```python
df=pd.DataFrame({'name':name,
                'rating':rating,
                'review':review,
                'salaries':salaries,
                'interviews':interviews,
                'jobs':jobs,
                'benifits':benifits,
                'photos':photos,
                'about':about})
```

```python
div_tag = soup.find('div', class_='td-post-content tagdiv-type')

# Check if the <div> tag exists and contains a <pre> tag
if div_tag:
    pre_tag = div_tag.find('pre')
    if pre_tag:
        # Remove the <pre> tag
        pre_tag.extract()
```

```python
div_tag = soup.find('div', class_='td-post-content tagdiv-type')

# Check if the <div> tag exists
if div_tag:
    # Extract all the text from the <div> tag excluding the text within <pre> tags
    text = div_tag.get_text(separator='\n', strip=True, exclude=['pre'])

    # Print or use the extracted text
    print(text)
```

```python
div_tag = soup.find('div', class_='page-cover')

# Check if the <div> tag exists
if div_tag:
    # Remove the text within <pre> tags
    for pre_tag in div_tag.find_all('pre'):
        pre_tag.extract()

    # Extract all the text from the <div> tag
    text = div_tag.get_text(separator='\n', strip=True)

    # Print or use the extracted text
    print(text)
```
